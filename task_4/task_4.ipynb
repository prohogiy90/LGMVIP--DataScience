```python
path = "/content/drive/MyDrive/Lets grow more/lgm2.jpg"
```

```python
# Importing Libraries

import cv2
from google.colab.patches import cv2_imshow
```

```python
# Load image

image = cv2.imread(path)
cv2_imshow(image)
```

```python
# Convert color space to gray

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# Inverting a Grayscale Image to a Negative

inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

```python
# Image Smoothing using OpenCV Gaussian Blur

blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
# Invert the blurred image and perform per-element division of two arrays to get a "pencil drawing"

inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256)
cv2_imshow(pencil_sketch)
```

```python
# Output both images - the original and after converting to a "pencil drawing"

cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
